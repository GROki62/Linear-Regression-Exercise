```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import statsmodels.api as sm

df = pd.read_csv('Regression_Dataset.csv')
df.head()
```

```python
df.info()
```

```python
Kids = df['KIDS']
Foodaway = df['FOODAWAY']

median_data = df[['FOODAWAY','KIDS']]

plt.figure(figsize = (20,20))
plt.ylabel('Amount of Money Spent on Food Away from Home')
plt.xlabel('Number of Kids in Family')
plt.title('Money Spent on Food vs Number of Kids in Family')
plt.scatter(Kids, Foodaway)
```

```python
x_ary = df['KIDS']
x_with_cnst = sm.add_constant(x_ary)
model = sm.OLS(df['FOODAWAY'], x_with_cnst)
results = model.fit()
b,m = results.params
print(m)
print(b)
```

```python
print(results.summary())
```

```python
plt.figure(figsize = (20,20))
plt.scatter(df['KIDS'], df['FOODAWAY'])
plt.plot(x_ary, m*x_ary+b, color='red')
```
